#### Here there is lots of zeros values in budget and revenues column ,  I will use the mean value to replace the zeros.

```python
# Get average of budget
budget_mean = df['budget'].mean()
# Replace 0 values with mean.
df['budget'] = df['budget'].replace(0, budget_mean)
# Get average of revenue
revenue_mean = df['revenue'].mean()
# Replace 0 values with mean
df['revenue'] = df['revenue'].replace(0, revenue_mean)

# Get average of budget_adj
budget_adj_mean = df['budget_adj'].mean()
# Replace 0 values with mean.
df['budget_adj'] = df['budget_adj'].replace(0, budget_adj_mean)
# Get average of revenue_adj
revenue_adj_mean = df['revenue_adj'].mean()
# Replace 0 values with mean
df['revenue_adj'] = df['revenue_adj'].replace(0, revenue_adj_mean)
df.head(1000)
```

# Q1 What is the relationship between runtime and and vote average
### in this qoustion i will use scatter plot to explain the relataion between run time and vote average , because it clearfy well the correlation between variables.

```python
#create a scatter plot to show the relationship between runtime and vote average
df.plot(x='runtime', y='vote_average', kind='scatter', figsize=(14,5))
plt.title('vote_average vs. runtime')
plt.xlabel('runtime')
plt.ylabel('vote_average');
```

# Q2 :Is more budget lead to good ratings?
### here also the scatter plot is usefull for relaionship between variables

```python
#Also here I created scatter plot to show the relation between the budget and the average vote 
plt.figure(figsize = (18, 10))
plt.title('Average Budget and Vote Average')
plt.xlabel("Budget (In 100,000,000's)")
plt.ylabel('Vote Average')
graph = plt.scatter(df['budget_adj'], df['vote_average'])
# here is the line to show the slop of the corrilation 
plt.plot(df['budget_adj'], np.poly1d(np.polyfit(df['budget_adj'], df['vote_average'], 1))(df['budget_adj']), color = 'black')

plt.show()
```

# Q3 :How is the yearly revenue change relative to the budget ?
### here I calculate the grwoth in revenue and then I used two types of charts first one to is for grwoth and the second to compare revenues and budget over years

```python
# to calculte the grwoth for revenue
def growth(mean_revenue):
    return mean_revenue - mean_revenue.shift(1).fillna(0)
```

```python
# here is the mean of revenue over years, include only movies which have revenue data
movies_budget = df[df.budget_adj > 0]
movies_revenue = movies_budget[movies_budget.revenue_adj > 0]

revenues_by_years = movies_revenue.groupby('release_year').sum()

revenues_by_years.apply(growth)['revenue'].plot()
plt.title('Revenue growth')
plt.ylabel('US dollar');
revenues_by_years[['budget_adj', 'revenue_adj']].plot()
plt.title('Revenue vs budget')
plt.ylabel('US dollar');
```

# Q4: How is the distribtuon of ratings?
### here I used this chart to show how the ratings are distributed and it tell that it 90% of data between 5 and 8

```python
# shows the average distrubtion ratings of all moveies 
plt.figure(figsize=(8, 6))
sns.kdeplot(df['vote_average'], shade=True, color="b")
plt.title('Average Rating Distribution', fontdict={'fontsize': 17})
plt.xlabel('Average Rating', fontdict={'fontsize': 14})
plt.ylabel('Relative Amount', fontdict={'fontsize': 14});
```
